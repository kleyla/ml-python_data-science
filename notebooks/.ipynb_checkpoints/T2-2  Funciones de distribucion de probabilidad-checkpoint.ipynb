# Funciones de distribucion de probabilidades

## Distribucion Uniforme

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
b = 100
n = 1000
data = np.random.uniform(a, b, n)
```

```python
%matplotlib inline
```

```python
plt.hist(data)
```

## Distribucion Normal

### La campana de Gauss

```python
data = np.random.randn(1000)
```

```python
x = range(1,1001)
plt.plot(x, data)
```

```python
%matplotlib inline
plt.hist(data)
```

```python
# sorted ordenada caracteriza una funcion gausiana
plt.plot(x, sorted(data))
```

```python
mu = 5.5
sd = 2.5
Z = np.random.randn(1000)
data = mu + sd*Z # Z= (X - mu) / sd
plt.hist(data)
```
