### Scatter plot

```python
# como los datos estan relacionados entre una variable y otra
data.plot(kind="scatter", x="Day Mins", y="Day Charge")
```

```python
data.plot(kind="scatter", x="Night Mins", y="Night Charge")
```

```python
figure, axs = plt.subplots(2,2, sharey=True, sharex=True)
#subplot que queda dentro de la matriz
data.plot(kind="scatter", x="Day Mins", y="Day Charge", ax=axs[0][0])
data.plot(kind="scatter", x="Night Mins", y="Night Charge", ax=axs[0][1])
data.plot(kind="scatter", x="Day Calls", y="Day Charge", ax=axs[1][0])
data.plot(kind="scatter", x="Night Calls", y="Night Charge", ax=axs[1][1])
```

## Histogramas de frecuencia

Donde se concentran los valores capmana de gaus

```python
# bins= las divisiones de los datos
# plt.hist(data["Day Calls"], bins=20)
# plt.hist(data["Day Calls"], bins=[0,60,90,120,150,180])

# Como dividir los datos?
# Regla de Sturges
# ceil redondea hacia arriba
k = int(np.ceil(1+np.log2(3333)))
plt.hist(data["Day Calls"], bins=k)

plt.xlabel("Nuemero de llamadas al dia")
plt.ylabel("Frecuencia")
plt.title("Histograma de numero de llamadas al dia")
```

## Boxplot, diagrama de caja y bigotes

```python
plt.boxplot(data["Day Calls"])
plt.ylabel("Numero de llamadas diarias")
plt.title("Boxplot de las llamadas diarias")
```

la caja indica donde estan la mayoria de datos:
* mediana es la que esta al medio (la linea naranja)
* moda
* promedio

```python
data["Day Calls"].sum()
```

```python
data["Day Calls"].mean()
```

```python
#cuantiles principales 25% 50% 75%
# de la caja el borde inferior seria 25% que es 87
# la mediana es el borde central naranja 50% que es 101
# el cuantil 75% cae en 114
data["Day Calls"].describe()
```

```python
# rango inter quartilico = tamanho de la caja 
# IQR = 114 - 87
IQR = data["Day Calls"].quantile(0.75)- data["Day Calls"].quantile(0.25)
IQR
```

```python
# Bigote Donde acaba el bigote = es la linea que esta abajo de la caja
data["Day Calls"].quantile(0.25) - 1.5*IQR
```

```python
#Bigote superior = linea que esta arriba de la caja
data["Day Calls"].quantile(0.75) + 1.5*IQR
```
